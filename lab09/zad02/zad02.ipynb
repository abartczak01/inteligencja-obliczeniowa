```python
import matplotlib.pyplot as plt
import random
from aco import AntColony

plt.style.use("dark_background")

# 20 wierchołków z losowymi parametrami od 0 do 100
COORDS = [(random.randint(0, 100), random.randint(0, 100)) for _ in range(20)]

def plot_nodes(coords, w=12, h=8):
    for x, y in coords:
        plt.plot(x, y, "g.", markersize=15)
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches([w, h])

def plot_all_edges(coords):
    paths = ((a, b) for a in coords for b in coords)
    for a, b in paths:
        plt.plot((a[0], b[0]), (a[1], b[1]))


plot_nodes(COORDS)
```

```python
colony = AntColony(COORDS, ant_count=300, alpha=0.5, beta=1.2, 
                    pheromone_evaporation_rate=0.40, pheromone_constant=1000.0,
                    iterations=300)

optimal_nodes = colony.get_path()

for i in range(len(optimal_nodes) - 1):
    plt.plot(
        (optimal_nodes[i][0], optimal_nodes[i + 1][0]),
        (optimal_nodes[i][1], optimal_nodes[i + 1][1]),
    )

plt.show()
```

```python
def plot_path(colony):
    optimal_nodes = colony.get_path()
    for i in range(len(optimal_nodes) - 1):
        plt.plot(
            (optimal_nodes[i][0], optimal_nodes[i + 1][0]),
            (optimal_nodes[i][1], optimal_nodes[i + 1][1]),
        )

    plt.show()
```

```python
colony_fewer_ants = AntColony(COORDS, ant_count=100, alpha=0.5, beta=1.2, 
                              pheromone_evaporation_rate=0.40, pheromone_constant=1000.0,
                              iterations=300)

plot_path(colony_fewer_ants)
```

```python
# 2. zmniejszenie iterations
colony_less_iterations = AntColony(COORDS, ant_count=300, alpha=0.5, beta=1.2, 
                                   pheromone_evaporation_rate=0.40, pheromone_constant=1000.0,
                                   iterations=100)

plot_path(colony_less_iterations)
```

```python
# 3. zwiekszenie pheromone_evaporation_rate
colony_high_evaporation = AntColony(COORDS, ant_count=300, alpha=0.5, beta=1.2, 
                                    pheromone_evaporation_rate=0.80, pheromone_constant=1000.0,
                                    iterations=300)

plot_path(colony_high_evaporation)
```

```python
colony_high_beta = AntColony(COORDS, ant_count=300, alpha=0.5, beta=2.0, 
                             pheromone_evaporation_rate=0.40, pheromone_constant=1000.0,
                             iterations=300)

plot_path(colony_high_beta)
```

1. Zmniejszenie liczby mrówek spowodowało dużo szybsze znalezienie rozwiązania, które jest tak samo dobre jak oryginał (395)
2. Zmniejszenie liczby itereacji - dużo szybsze znalezienie rozwiązania, gorsze rozwiązanie
3. zwiekszenie x2 pheromone_evaporation_rate - nieco szybsze znalezeienie rozwiazania, gorsze rozwiązanie
4. zwiekszenie beta - nieco szybsze znalezienie rozwiazania, tak samo dobre rozwiazanie
